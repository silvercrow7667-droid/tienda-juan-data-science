#1. Análisis de facturación



```python
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print(f"El ingreso total de la tienda 1 es de: ${ingreso_tienda:,.2f}")
print(f"El ingreso total de la tienda 2 es de: ${ingreso_tienda2:,.2f}")
print(f"El ingreso total de la tienda 3 es de: ${ingreso_tienda3:,.2f}")
print(f"El ingreso total de la tienda 4 es de: ${ingreso_tienda4:,.2f}")
```

#📄Informe final

##Introducción

El presente análisis tiene como objetivo ayudar al Sr. Juan a decidir qué tienda de la cadena Alura Store debería vender para iniciar un nuevo emprendimiento. Para ello, se analizaron los datos de ventas de cuatro tiendas, considerando métricas clave como los ingresos totales, la distribución de ventas por categoría, la calificación promedio de los clientes, los productos más y menos vendidos y el costo de envío promedio.
Además, se utilizaron visualizaciones de datos para facilitar la interpretación de los resultados y apoyar la toma de decisiones.

##Análisis de resultados

Ingresos totales
El análisis de facturación muestra que la Tienda 1 presenta el mayor ingreso total, seguida por la Tienda 2 y la Tienda 3. La Tienda 4 registra el ingreso más bajo entre todas, lo que indica un menor desempeño financiero general en comparación con las demás.

Ventas por categoría
Al analizar las ventas por categoría, se observa que en todas las tiendas las categorías de Electrónicos y Muebles se encuentran entre las más vendidas. Sin embargo, la Tienda 4 no destaca significativamente en ninguna categoría específica, mientras que otras tiendas muestran un mejor equilibrio y mayor volumen de ventas en categorías clave.

Calificación promedio de los clientes
Las calificaciones promedio de los clientes son relativamente similares entre las tiendas, con valores cercanos a 4 puntos. La Tienda 3 presenta la mejor calificación promedio, seguida de la Tienda 2. La Tienda 4, aunque mantiene una buena valoración, no logra diferenciarse de forma positiva frente a las demás.

Productos más y menos vendidos
Cada tienda cuenta con productos destacados en ventas; sin embargo, la Tienda 4 presenta una diferencia marcada entre su producto más vendido y el menos vendido, lo que puede indicar una menor diversidad de productos exitosos o una oferta menos equilibrada en comparación con otras tiendas.

Costo de envío promedio
El costo de envío promedio es similar entre las cuatro tiendas, sin diferencias significativas que representen una ventaja competitiva clara para alguna de ellas. Por lo tanto, este factor no compensa el menor ingreso observado en la Tienda 4.

##Conclusión y recomendación

Con base en el análisis de los ingresos, las ventas por categoría, la satisfacción del cliente, el desempeño de los productos y el costo de envío, se concluye que la Tienda 4 es la opción más adecuada para ser vendida.

A pesar de mantener una calificación promedio aceptable por parte de los clientes, la Tienda 4 presenta el menor ingreso total, no destaca en categorías de alto volumen de ventas y no muestra ventajas claras en términos de costos de envío o desempeño de productos frente a las demás tiendas. En contraste, las otras tiendas exhiben mejores resultados financieros y un mayor potencial de crecimiento.

Por lo tanto, se recomienda que el Sr. Juan venda la Tienda 4, ya que su venta tendría un menor impacto negativo en el rendimiento general de la cadena y permitiría destinar recursos a un nuevo emprendimiento con mayor potencial de rentabilidad.
